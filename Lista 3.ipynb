## Problema 1

```python
#a)
v = 100
N = 100
L = 1
a = L/N
d = 0.1
C = 1
sig = 0.3
h = 1e-6
eps = h/1000

x = linspace(0, L, N)

phi = zeros(N, float)
phinovo = copy(phi)

psi = zeros(N, float)
```

```python
def fpsi(x):
    temp = (x*L-x**2)*C/L #coloquei apenas L, pois ele Ã© igual a 1
    temp2 = exp(-((x-d)**2)/(2*(sig**2)))
    
    return temp*temp2

for i in range(N):
    psi[i] = fpsi(x[i])
    
psinovo = copy(psi)
```

```python
#b)
t1, t2, t3, t4, t5, t6, t7 =  1e-4, 1e-3, 2e-3, 1e-2, 0.1, 0.2, 0.3
t = 0.0
while (t<(t7+eps)):
    
    for i in range(1, N-1):
        phinovo[i] = phi[i] + h*psi[i]
        psinovo[i] = psi[i] + h*(phi[i-1]+phi[i+1]-2*phi[i])*(v**2)/(a**2)
    t+=h
    phi = copy(phinovo)
    psi = copy(psinovo)
    if (abs(t-t1)<=eps) or (abs(t-t2)<=eps) or (abs(t-t3)<=eps) or (abs(t-t4)<=eps) or (abs(t-t5)<=eps) or (abs(t-t6)<=eps) or (abs(t-t7)<=eps):
        figure(figsize=(8,5))
        plot(x,phi,label="t = %.4f" % t)
        legend()
```

## Problema 2

```python
from numpy import max
```

```python
#a)
m = 1
g = 9.81
N = 100
t = linspace(0, 10, N)
h = 10/N
erro = 10
tol = 1e-6

x = zeros(N, float)
xnovo = copy(x)
```

```python
#b)
while (abs(erro)>tol):
    for i in range(1,N-1):
        xnovo[i] = (x[i+1] + x[i-1] + g*(h**2))/2
        
    erro = max(abs(xnovo - x))
    x = copy(xnovo)
```

```python
figure(figsize=(8,5))
plot(t, x)
xlabel("t")
ylabel("x(t)")
title("TrajetÃ³ria da Bola")
```
