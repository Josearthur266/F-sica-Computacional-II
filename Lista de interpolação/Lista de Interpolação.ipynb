```python
from numpy import*
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

## Problema 1

```python
#a)
dados = loadtxt('scattering.data', float)
x = dados[:,0]
y = dados[:,1]
```

```python
#b)
f1 = interp1d(x, y)
```

```python
#c)
t = linspace(0, 200, 200)
plt.plot(f1(t))
plt.plot(x, y, 'o')
```

```python
#d)
f2 = interp1d(x, y, kind = "cubic")
plt.plot(f2(t))
plt.plot(x, y, 'o')
```

## Problema 2

```python
#a)
def f(t, i):
    return (y[i-1]*(x[i]-t) + y[i]*(t - x[i-1]))/(x[i] - x[i - 1])
```

```python
#b)
t = []
v = []
for i in range(1, len(x)):
    temp = linspace(x[i-1], x[i], 25)
    for q in range(len(temp)):
        v.append(f(temp[q], i))
        t.append(temp[q])
        
plt.plot(t, v)
plt.plot(x, y, 'o')
```

## Problema 3

```python
#a)
def fg1(t):
    g = 0
    for i in range(9):
        gamma = 1
        for j in range(9):
            if j != i:
                gamma = gamma*((t-x[j])/(x[i]-x[j]))
        g += y[i]*gamma
    return g
```
